```python
import pandas as pd

# 1. ファイルの読み込み
# 0.897を出した最強ファイル（ファイル名は適宜変更してください）
best_csv = pd.read_csv('/kaggle/input/20260207/submission5.csv') 

# 今作っているSpectrogramモデルのファイル
spec_csv = pd.read_csv('/kaggle/input/20260207/submission_spectrogram_2dcnn.csv')

# 2. アンサンブル（加重平均）
# CNN-LSTM (0.897) は信頼度が高いので 70%、Spectrogram は 30% 混ぜます
# 異なる視点（波形 vs 画像）を混ぜることで、お互いのミスを打ち消します
submission = best_csv.copy()
submission['lever'] = (best_csv['lever'] * 0.6) + (spec_csv['lever'] * 0.4)

# 3. 保存
submission.to_csv('submission_ensemble_06_04.csv', index=False)
print("Ensemble Done! Saved to submission_ensemble_06_04.csv")
```
